```python
from sklearn.metrics import f1_score
```

```python
#Load final data
final_df = pd.read_csv("data_3/results.csv")
final_df.head()

f1_macro_olid_small_CNN = f1_score(final_df.labels, final_df.CNN_olid_small, average='macro')
f1_macro_hasoc_CNN = f1_score(final_df.labels, final_df.CNN_hasoc, average='macro')
olid_small_CNN_error = np.std(final_df.CNN_olid_small)
hasoc_CNN_error = np.std(final_df.CNN_hasoc)

f1_macro_olid_small_BiLSTM = f1_score(final_df.labels, final_df.BiLSTM_olid_small, average='macro')
f1_macro_hasoc_BiLSTM = f1_score(final_df.labels, final_df.BiLSTM_hasoc, average='macro')
olid_small_BiLSTM_error = np.std(final_df.BiLSTM_olid_small)
hasoc_BiLSTM_error = np.std(final_df.BiLSTM_hasoc)

f1_macro_olid_small_roberta = f1_score(final_df.labels, final_df.roberta_olid_small, average='macro')
f1_macro_hasoc_roberta = f1_score(final_df.labels, final_df.roberta_hasoc, average='macro')
olid_small_roberta_error = np.std(final_df.roberta_olid_small)
hasoc_roberta_error = np.std(final_df.roberta_hasoc)

f1_macro_olid_small_hateBERT = f1_score(final_df.labels, final_df.hateBERT_olid_small, average='macro')
f1_macro_hasoc_hateBERT = f1_score(final_df.labels, final_df.hateBERT_hasoc, average='macro')
olid_small_hateBERT_error = np.std(final_df.hateBERT_olid_small)
hasoc_hateBERT_error = np.std(final_df.hateBERT_hasoc)

f1_macro_olid_small_NB = f1_score(final_df.labels, final_df.NB_olid_small, average='macro')
f1_macro_hasoc_NB = f1_score(final_df.labels, final_df.NB_hasoc, average='macro')
olid_small_NB_error = np.std(final_df.NB_olid_small)
hasoc_NB_error = np.std(final_df.NB_hasoc)

f1_macro_olid_small_SVM = f1_score(final_df.labels, final_df.SVM_olid_small, average='macro')
f1_macro_hasoc_SVM = f1_score(final_df.labels, final_df.SVM_hasoc, average='macro')
olid_small_SVM_error = np.std(final_df.SVM_olid_small)
hasoc_SVM_error = np.std(final_df.SVM_hasoc)
```

```python
from numpy import size


SMALL_SIZE = 10
MEDIUM_SIZE = 12
BIGGER_SIZE = 14

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

labels = ["CNN", "BiLSTM", "Roberta", "hateBERT", "NB", "SVM"] 
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

olid_small = [f1_macro_olid_small_CNN, f1_macro_olid_small_BiLSTM, f1_macro_olid_small_roberta, f1_macro_olid_small_hateBERT, f1_macro_olid_small_NB, f1_macro_olid_small_SVM]
hasoc = [f1_macro_hasoc_CNN, f1_macro_hasoc_BiLSTM, f1_macro_hasoc_roberta, f1_macro_hasoc_hateBERT, f1_macro_hasoc_NB, f1_macro_hasoc_SVM]

fig, ax = plt.subplots()
fig.set_figheight(10)
fig.set_figwidth(10)
rects1 = ax.bar(x - width/2, olid_small, width, label='Trained on Olid small',  align='center', alpha=0.5, ecolor='black', capsize=10)
rects2 = ax.bar(x + width/2, hasoc ,width, label='Trained on Hasoc', align='center', alpha=0.5, ecolor='black', capsize=10)
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('F1 macro')
ax.set_title('Macro Average F1-score on the olid test set', fontsize = BIGGER_SIZE)
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```
